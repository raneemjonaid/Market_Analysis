### Step 3.2: Fill Missing Brands Using First Word
Fill missing brands where the first word of the title matches a known brand.

```python
# Roll back and use ONLY Method 1
data_filled = data.copy()

# Method 1 ONLY: First word extraction
for idx, row in data_filled[data_filled['brand'].isnull()].iterrows():
    first_word = row['title'].split()[0]
    if first_word in known_brands:
        data_filled.loc[idx, 'brand'] = first_word

print(f"Missing brands before: {data['brand'].isnull().sum()}")
print(f"Missing brands after Method 1: {data_filled['brand'].isnull().sum()}")
print(f"Brands filled: {data['brand'].isnull().sum() - data_filled['brand'].isnull().sum()}")
```


## Step 4: Handling Duplicates
Before handling missing values, we first check if any products 
appear more than once in the dataset. If a duplicate has a value 
that another duplicate is missing, we can use it to fill the gap.

### Step 4.1: Check for Duplicates
We check how many duplicate rows exist in the dataset.
A duplicate row is an exact copy of another row.

```python
# Check for duplicate products
print("Total rows:", len(data_filled))
print("Duplicate rows:", data_filled.duplicated().sum())
print()

```

### Step 4.2: Count How Many Times Each Product Appears
We group duplicate products by title and count their occurrences
to understand how many times each product is repeated.

```python
# Count occurrences of each title
title_counts = data_filled[data_filled.duplicated(subset=['title'], keep=False)].groupby('title').size().reset_index()
title_counts.columns = ['title', 'count']
title_counts = title_counts.sort_values('count', ascending=False)
print(title_counts)

```

### Step 4.3: Fill Missing Values Using Duplicate Products
If a product appears multiple times and one entry has a missing value 
while another entry has the actual value, we fill the missing value 
from its duplicate.

#### Columns we attempt to fill:
- price
- rating
- reviews
- sold_past_month

```python
# Fill missing values using duplicate products
cols_to_fill = ['price', 'rating', 'reviews', 'sold_past_month']

for col in cols_to_fill:
    # Group by title and fill NaN with first available value
    data_filled[col] = data_filled.groupby('title')[col].transform(
        lambda x: x.fillna(x.dropna().iloc[0] if len(x.dropna()) > 0 else x.iloc[0])
    )

# Check results
print("Missing values after filling from duplicates:")
print(data_filled[cols_to_fill].isnull().sum())
```

### Step 5.1: Drop Rows with Missing Price
Price is essential for our analysis. 
We drop the 65 rows where price is missing.

```python
# dropna() removes rows where price is NaN
# subset=['price'] tells it to only check the price column
before = len(data_filled)
data_filled = data_filled.dropna(subset=['price'])
after = len(data_filled)

print(f"Rows before: {before}")
print(f"Rows after: {after}")
print(f"Rows dropped: {before - after}")
```

### Step 5.2: Remove Duplicate Rows
Keep only one copy of each duplicated product.

```python
# drop_duplicates() removes rows with same title
# keeping only the first occurrence
before = len(data_filled)
data_filled = data_filled.drop_duplicates(subset=['title'])
after = len(data_filled)

print(f"Rows before: {before}")
print(f"Rows after: {after}")
print(f"Duplicates removed: {before - after}")
```

Step 5.3 - Verify final dataset 


```python
# Final check to confirm our dataset is clean
print(f"Final dataset size: {len(data_filled)}")
print()
print("Missing values:")
print(data_filled.isnull().sum())
print()
print("Duplicates remaining:", data_filled.duplicated().sum())
```

### Step 6.1: Fix Data Types
Convert price, rating, reviews and sold_past_month from text to numbers.

```python
# price: convert to float
data_filled['price'] = pd.to_numeric(data_filled['price'], errors='coerce')

# reviews: remove commas and convert to number
data_filled['reviews'] = data_filled['reviews'].str.replace(',', '').str.strip()
data_filled['reviews'] = pd.to_numeric(data_filled['reviews'], errors='coerce')

# Check results
print("Data types after fixing:")
print(data_filled[['price', 'rating', 'reviews']].dtypes)
print()
print("Sample values:")
print(data_filled[['price', 'rating', 'reviews']].head())
```

```python
# Check current data types and sample values
print("Current data types:")
print(data_filled[['price', 'rating', 'reviews', 'sold_past_month']].dtypes)
print()
print("Sample values:")
print(data_filled[['price', 'rating', 'reviews', 'sold_past_month']].head())
```

### Step 6.2: Fix sold_past_month Column
Convert sold_past_month from text to numbers.
- "1K+" becomes 1000
- "8K+" becomes 8000
- "100+" becomes 100

```python
# Convert sold_past_month to numbers
# 1K+ = 1000, 8K+ = 8000, 100+ = 100
def convert_sold(value):
    if pd.isna(value):
        return None
    value = str(value).replace('+', '').strip()
    if 'K' in value:
        return float(value.replace('K', '')) * 1000
    else:
        return float(value)

data_filled['sold_past_month'] = data_filled['sold_past_month'].apply(convert_sold)

print("Data types now:")
print(data_filled[['price', 'rating', 'reviews', 'sold_past_month']].dtypes)
print()
print("Sample values:")
print(data_filled[['price', 'rating', 'reviews', 'sold_past_month']].head())
```
